#### Présentation des fiches

Représentation plus organisées :

- cartes perforées !

- Une fiche par ami.
- sur la fiche : 
  - la liste des recettes connues
  - en face de chaque recette un trou
  - lorsque **l'ami en question aime cette recette**, on découpe le papier entre le trou et le bord de la carte 

Et maintenant, pour poser des questions on a besoin de baguettes.
