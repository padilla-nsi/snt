## Web et Internet

### Internet, un réseau informatique

* Réseau informatique. Un *réseau informatique* est un ensemble de nœuds (PC, mobiles . . . ) qui peuvent communiquer entre eux par des liens de natures diverses (fils, cables sous-marins, ondes . . . ).
* Internet. *Internet* désigne une interconnexion mondiale de réseaux d’ordinateurs.
* Adresse IP. Chaque nœud doit disposer d’une adresse unique sur le réseau pour communiquer avec les autres. Sur Internet il s’agit d’une *adresse IP*.
* TCP/IP. Des règles universelles, appelées *protocoles* permettent aux nœuds de réseaux basés sur des technologies hétérogènes (Ethernet, Wifi, 4G . . . ) d’échanger sur Internet. Sur Internet il s’agit de la pile de protocoles d’une *TCP/IP*.

### Naissance d'Internet


* 1960. Les premières expériences de connexions d’ordinateurs en réseaux ont lieu dans les années 60.
* 1969. Le 29 octobre 1969 a lieu par exemple la première connexion entre des ordinateurs d’universités distinctes : Stanford et UCLA.

* 1970. Dans les années 1970, Vinton Cerf et Robert Kahn, s’inspirant des travaux de Louis Pouzin développent la pile de protocole TCP/IP.
* 1983. À partir de 1983, TCP/IP permet d’interconnecter les réseaux ARPANET et CSNET donnant naissance à l’Internet. Internet reste principalement utilisé par des chercheurs et des
acteurs institutionnels.

### Le Web, un réseau de documents

* Web. Le *Web* est un réseau de documents reliés par des liens hypertextes.
* Client/serveur. Le Web est un service d’Internet basé sur une architecture *client/serveur* :
  * un serveur fournit les ressources qui lui sont demandées par des clients
  * un logiciel client est un navigateur (par exemple Firefox, Chrome, Edge, Safari...)

![](04-client-serveur.png)

* HTTP. Clients et serveurs Web communiquent avec le protocole *HyperText Transfer Protocol* (HTTP).
* URL. Chaque ressource du Web a une adresse unique appelée *Uniform Ressource Locator* (URL).

### Naissance du Web

* 1989. Tim Berners-Lee, qui s’occupe de la gestion des documents scientifiques au CERN, propose avec le projet mesh (filet) de faire traverser le réseau internet à l’*hypertexte* imaginé par Ted Nelson en 1965.

![](05-premiere-page-web.png)


* 1993. Les dirigeants du CERN annoncent officiellement que la technologie du Web sera gratuite et libre de droits. Microsoft essaie de développer des formats propriétaires concurrents mais en vain (la première version de Windows 95 était livrée sans navigateur Web !)
* 1994. Le système des *cookies* petits fichiers textes échangés entre client set serveur, est intégré dans le navigateur Netscape afin de conserver des informations sur le client (panier) lors de sa navigation sur un site Web.
* 2000. Le chiffre d’affaire de la publicité sur le Web est de 8 milliards d’euros.
* Les années 2000 sont celles du Web 2.0 ou Web social avec l’apparition des blogs, des réseaux sociaux, des plateformes et l’extension du Web à la mobilité avec le succès des smartphones.
