# Le routage sur Internet

Comment un e-mail ne comportant que l'adresse de l'**expéditeur** et du **destinataire** fait-il pour circuler sur Internet.... et arriver à destination ?

Nous allons essayer de comprendre cela !

## 1. Mise en place

**Étape 1.**
Chaque élève pioche une adresse email qui doit rester **secrète** (parmi `eleve_1@lycee.com`, `eleve_2@lycee.com`,... , `eleve_36@lycee.com`, etc.)

**Étape 2.**
Chaque élève prend connaissance de son **correspondant** : la personnes à laquelle elle peut écrire.

* Djamel ~(eleve2@lycee.com)~ -> Ada ~(eleve_1@lycee.com)~
* Frédérique ~(eleve_3@lycee.com)~ -> Djamel ~(eleve2@lycee.com)~

## 2. Premier message

**Étape 3.**
Chaque élève prend une demi-feuilles de papiers (le **mail**) et écrit dessus :


    expéditeur: <mail_secret@@lycee.com>
    sujet: Hello
    <mon prénom>
    

**Étape 4.**
Chaque élève envoie son mail: il donne la feuille à son correspondant.

Quelles informations nouvelles possède chaque correspondant ? Quel est maintenant le nombre de correspondant de chaque élève ?

## 3. Premières réponses

**Étape 5.**
Chaque élève note sur une demi-feuille personnelle sa table de routage:

    <son prénom> (par -)
    <prénom du correspondant> (par <mail du correspondant>)


Que note Djamel dans sa table de routage ? Et Ada ?

**Étape 6.**
Chaque élève envoie à tous ses correspondants une copie de sa table de routage.

Est ce que les correspondants reçoivent de nouvelles informations ? Si oui, lesquelles ?

## 4. Mise à jour des tables de routage

**Étape 7.**
Lorsqu'un élève reçoit la table de routage d'un de ses correspondant, il met à jour, si besoin, sa tables de routage : 

* il ajoute une ligne pour chaque **nouveau prénom** et y écrit:

    <prénom nouveau> (par <mail du correspondant>)

Q1. Compléter les tables de Djamel et d'Ada.

Q2. Est-ce que la table de Frédérique est modifiée ? Si oui, écrire sa table.

**Étape 8.** Tant qu'un élève a sa table de routage modifiée par une mise à jour, on reprend à l'étape 6.

Écrire les tables finales obtenues par Djamel, Ada et Frédérique lorsqu'il n'y a plus de mail échangés.
